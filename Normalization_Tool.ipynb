```
from sklearn import preprocessing
import numpy as np
import pandas as pd

file_name = "/content/bq-results-20200601-201956-fhtrmoarf9b5.csv"
# Get dataset
df = pd.read_csv(file_name, sep=",",dtype={'clientId':str})
# Normalize total_bedrooms column
df.describe()
```

```
print(len(df.columns))
print(df.columns)
no_of_column = len(df.columns)

```

```
no_of_feature = 1
test = df.iloc[:, list(range(0, no_of_feature))].astype(str)
print(test)
dictOffeature = { i-no_of_feature : df.columns[i] for i in range(no_of_feature ,len(df.columns)) }
print(dictOffeature) 
```

```
df.head()
```

```
x = df.iloc[:, list(range(no_of_feature, no_of_column))].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
xd = pd.DataFrame(x_scaled)
xd
```

```
xd.rename(columns= dictOffeature, 
                 inplace=True)
xd
```

```
result = pd.concat([test, xd], axis=1, sort=False)
result.head()
```

```
new_file_name = file_name.split("/")[-1].split('.')[0]
print(new_file_name)
result.to_csv('normalised_{}.csv'.format(new_file_name), index = False)
```
